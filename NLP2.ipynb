```python
import nltk
```

```python
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
sentences=["Although I love machine learning, I sometimes struggle with debugging complex models when the data is messy.",

    "Python is very powerful, and because it supports multiple libraries, developers can easily build scalable and efficient applications.",

    "This project is simple and fun, but if you add real-world constraints, such as limited memory and processing power, it becomes much more challenging."]

for sentence in sentences:
    words=nltk.word_tokenize(sentence)
    tags=nltk.pos_tag(words)

    print("Sentence:",sentence)
    print("Tags:",tags)

    for word, tag in tags:
        print(f"{word} - {tag}")
```
