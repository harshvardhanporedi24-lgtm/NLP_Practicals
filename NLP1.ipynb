```python
# Porter Stemmer: The oldest and very popular algorithm, known for its balance of accuracy and performance.
# Snowball Stemmer (or Porter2 Stemmer): An improved version of the Porter stemmer that supports many languages beyond English.
# Lancaster Stemmer: A more aggressive stemmer that can produce shorter, often non-dictionary, stems compared to the Porter stemmer.
# Regexp Stemmer: Allows you to define stemming rules using regular expressions.
# for stemming we use ps.stem(word) package-from nltk.stem import PorterStemmer
```

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

ps = PorterStemmer()

sentence = "Cars are wheeled motor vehicles for transporting people, featuring an engine, four wheels, and typically seating 1-8, powered by gasoline, diesel, or electricity."
words = word_tokenize(sentence)

stemmed_words = [ps.stem(word) for word in words]

print("Original words:", words)
print("Stemmed words:", stemmed_words)
```

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
sentence = "Cars are wheeled motor vehicles for transporting people, featuring an engine, four wheels, and typically seating 1-8, powered by gasoline, diesel, or electricity."
words = word_tokenize(sentence)
lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
print("Original words:", words)
print("Lemmatized words:", lemmatized_words)
```

```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem import PorterStemmer


stop_words = set(stopwords.words('english'))
sentence="""Cars are wheeled motor vehicles for transporting people, featuring an engine, four wheels, and typically seating 1-8, powered by gasoline, diesel, or electricity."""

words = word_tokenize(sentence)
filtered_words = [word for word in words if word.lower() not in stop_words]
print("Original words:", words)
print("Filtered words:", filtered_words)

```
