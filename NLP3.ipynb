```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
from nltk import ngrams
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
nltk.download('wordnet')
```

```python
sentence="My name is Harshvardhan. I am studying in Pimpri Chinchwad College of Engineering."
sentence=sentence.lower()
```

```python
tokens=word_tokenize(sentence)
stop_words=set(stopwords.words('english'))
filtered_sentence=[word for word in tokens if word not in stop_words]
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
lemmatized_sentence=[lemmatizer.lemmatize(word) for word in filtered_sentence]
print(lemmatized_sentence)
```

```python

bigrams=list(ngrams(lemmatized_sentence,2))
trigrams=list(ngrams(lemmatized_sentence,3))
print("Bigrams:",bigrams)
print("Trigrams:",trigrams)
```

```python
def get_ngrams(n, tokens):
    ngrams = []
    tokens = (n - 1) * ['<START>'] + tokens

    for i in range(n - 1, len(tokens)):
        context = tuple(tokens[i - p - 1] for p in reversed(range(n - 1)))
        target_word = tokens[i]
        ngrams.append((context, target_word))

    return ngrams
```

```python
sentence = "i love natural language processing"
tokens = sentence.split()

trigrams = get_ngrams(3, tokens)

for context, word in trigrams:
    print(f"{context} -> {word}")
```
